## Data loading

### Subtask:
Load the Kaggle API credentials from the "kaggle.json" file.


**Reasoning**:
Load the Kaggle API credentials from the "kaggle.json" file and store them in a dictionary.



```python
import json
import os

try:
    with open('kaggle.json', 'r') as f:
        kaggle_credentials = json.load(f)
except FileNotFoundError:
    print("Error: 'kaggle.json' not found. Please ensure the file exists in the current directory.")
    kaggle_credentials = None
except json.JSONDecodeError:
    print("Error: Invalid JSON format in 'kaggle.json'. Please check the file contents.")
    kaggle_credentials = None
```

## Data exploration

### Subtask:
Explore the loaded Kaggle API credentials.


**Reasoning**:
Explore the loaded Kaggle API credentials based on the provided instructions.



```python
if kaggle_credentials is None:
    print("Kaggle credentials could not be loaded. Exiting exploration.")
else:
    print("Keys in kaggle_credentials:", kaggle_credentials.keys())
    print("\nData types of credential values:")
    for key, value in kaggle_credentials.items():
        print(f"{key}: {type(value)}")

    print("\nExamination of credential values:")
    if 'username' in kaggle_credentials:
        print(f"Username: {kaggle_credentials['username']}")
    else:
        print("Username key not found.")

    if 'key' in kaggle_credentials:
        print(f"Key: {kaggle_credentials['key'][:5]}... (truncated for security)")  # Show first few characters
    else:
        print("Key key not found.")

    print("\nSummary of Kaggle Credentials:")
    for key, value in kaggle_credentials.items():
      print(f"- {key}: {value} (type: {type(value)})")

    if 'username' not in kaggle_credentials or 'key' not in kaggle_credentials:
        print("\nPotential issue: 'username' or 'key' is missing.")
    elif not kaggle_credentials['username'] or not kaggle_credentials['key']:
        print("\nPotential issue: 'username' or 'key' value is empty.")
    elif not isinstance(kaggle_credentials['username'], str) or not isinstance(kaggle_credentials['key'], str):
        print("\nPotential issue: 'username' or 'key' is not a string.")
```

```python
# Produtos mais exportados (Top 10)
top_produtos = df_exp.groupby("CO_NCM")["VL_FOB"].sum().sort_values(ascending=False).head(10)

# Mostra o resultado
top_produtos
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras dos produtos mais exportados
top_produtos.plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title("Top 10 Produtos Mais Exportados (por VL_FOB)")
plt.xlabel("Código NCM")
plt.ylabel("Valor FOB (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Agrupa por país e soma os valores FOB
top_paises = df_exp.groupby('CO_PAIS')['VL_FOB'].sum().sort_values(ascending=False).head(10)

# Junta com a tabela de nomes dos países
top_paises = top_paises.reset_index().merge(df_pais, on='CO_PAIS')

# Gráfico
plt.figure(figsize=(10,5))
plt.bar(top_paises['NO_PAIS'], top_paises['VL_FOB'], color='mediumseagreen')
plt.title("Top 10 Países que Mais Importam do Brasil")
plt.xlabel("País")
plt.ylabel("Valor FOB (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
